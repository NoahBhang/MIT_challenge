* **문자열(Strings)**: 문자들의 시퀀스로, 인덱싱과 슬라이싱(`my_string[1:3]`) 및 결합(`+`)이 가능합니다.

  * `while loop`: 특정 조건이 `True`인 동안 코드를 반복 실행합니다.

### 🏃‍♂️ Practice Problem 2
Write a program to remove the nth character from a non empty string. Print the old string and the new string.
(비어있지 않은 문자열에서 n번째 문자를 제거하고, 기존 문자열과 새 문자열을 각각 출력하세요.)


```python
test_string = "We want to remove the nth character from this string"
n = 8

# 학교 제공 답안 방식 연산 (인덱스 보정 처리)
new_string = test_string[:n-1] + test_string[n:]
print("Old:", test_string)
print("New:", new_string)
```

### 🏃‍♂️ Practice Problem 3
Does a given string have length greater than 10 or less than 5? If True, output True. If False, output False.
(주어진 문자열의 길이가 10보다 크거나 5보다 작으면 True, 그렇지 않으면 False를 출력하세요.)

```python
my_string = "This is my string"

if len(my_string) > 10 or len(my_string) < 5:
    print(True)
else:
    print(False)
```

### 🏃‍♂️ Practice Problem 4
Write a program which answers the following using a for loop: Count the number of e's in the following string.
(for 반복문을 사용하여 주어진 문자열 안에 소문자 'e'가 총 몇 개 들어있는지 개수를 세어 출력하세요.)

```python
my_string = "How many times is the letter e in this string?"
number_of_es = 0  # 카운터 변수 초기화

for letter in my_string:
    if letter == "e":
        number_of_es += 1

print(number_of_es)
```
