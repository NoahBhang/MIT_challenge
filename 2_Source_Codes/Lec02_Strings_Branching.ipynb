### 🧬 문자열 복합 연산과 형변환 (`int()`)
* **연산자 우선순위**: 문자열 연산에서도 곱셈(`*`)이 덧셈(`+`)보다 먼저 연산된다. `b + 2 * c`는 문자열 `c`를 두 번 반복한 뒤 `b`와 결합한다.
* **강제 형변환 (`int()`)**: 문자열 형태의 숫자 `"3"`은 그대로 곱하면 문자열이 반복되지만, `int(h)`를 사용해 정수 `3`으로 바꾸면 숫자로 활용되어 문자열을 원하는 횟수만큼 정확히 곱할 수 있게 된다.


```python
b = ":"
c = ")"
s1 = b+2*c
```

```python
f = "a"
g = " b"
h = "3"
s2 = (f+g)*int(h) #여기서 h = "3"인데 int(h)이라는 명령어를 사용하여 문자 3을 정수 3으로 '형변환' 한 것이다.
```

```python
print(s1)
print(s2)
print(s1)
```

### ✂️ 문자열 슬라이싱(Slicing) 기본 문법
* **기본 구조**: `s[start : stop : step]` 형식을 가진다. 이때 `stop` 인덱스 바로 직전까지만 가져온다는 점이 핵심이다.
  * `s[3:6]`: 인덱스 3부터 5까지 (`def`)
  * `s[3:6:2]`: 인덱스 3부터 5까지 2칸 간격으로 (`df`)
* **전체 출력 및 역순**:
  * `s[:]`: 시작과 끝을 생략하면 문자열 전체를 가져온다.
  * `s[::-1]`: `step`에 `-1`을 주면 문자열을 뒤에서부터 역순(`gfedcba`)으로 뒤집어 출력한다.


```python
#슬라이싱 기법은 모든 문자를 가져올 수도 있고, 한 칸씩 띠어서 가져올 수도 있고, 다른 문잘을 가져올 수도 있다.
# [start: stop: step]

s= "abcdefg"
s[3:6]
```

```python
s[3:6:2]
```

```python
s[:]
```

```python
s[: : -1]
```

### 🔍 문자열 슬라이싱의 고급 활용 및 주의점
* **음수 간격 슬라이싱**: `step`이 음수(`-2`)일 때는 거꾸로 이동한다. `s[4:1:-2]`는 인덱스 4부터 시작해 왼쪽으로 2칸씩 이동하며 인덱스 2까지의 문자를 가져온다.
* **동적 길이 활용 (`len()`)**: `len(s)`를 활용해 문자열의 길이를 구한 뒤 슬라이싱 범위를 지정할 수 있다.
* **빈 문자열 반환 예외**: `s[2:0]`이나 `s[6:3]`처럼 정방향(오른쪽)으로 전진해야 하는 범위인데 `start`가 `stop`보다 크면, 아무것도 찾지 못하고 빈 문자열(`''`)을 반환한다.


```python
s[4:1:-2]
```

```python
s = "ABC d3f ghi"
```

```python
s[3:len(s)-1]
```

```python
 s[4:0:-1]
```

```python
s[6:3]
```

### 📥 사용자 입력(`input()`)과 출력 제어
* **input()의 기본 타입**: `input()` 함수로 입력받은 데이터는 숫자를 타이핑하더라도 **무조건 문자열(String)**로 인식됩니다. 그렇기 때문에 `5 * num1`을 하면 곱셈이 아니라 입력한 문자가 5번 반복 출력된다.
* **f-string (포맷 스트링)**: 문자열 앞에 `f`를 붙이고 `{변수명}`을 넣으면 문장 중간에 변수 값을 가장 깔끔하고 직관적으로 삽입할 수 있다.
* **공백 제어 테크닉**: 문장 마지막에 생기는 불필요한 공백을 제거하기 위해 `(question + ' ') * 4 + question`과 같이 구조적으로 연산식을 설계할 수 있다.


```python
num1 = input('Type anything: ')
print(5*num1) # 숫자 3을 입력했으나 15가 아닌, 문자 5를 다섯 번 출력한다.
```

### 🎮 입력 정수 변환과 조건문(`if-elif-else`)을 활용한 숫자 맞추기 게임
* **입력 즉시 형변환**: 숫자로 연산을 해야 할 때는 `int(input())`과 같이 입력받자마자 정수로 바꾸어 변수에 저장해야 정상적인 산술 연산(`5 * num2 = 15`)이 가능하다.
* **조건문 흐름 제어**: `if`, `elif`, `else` 구조를 사용하여 입력된 값(`guess`)과 정답(`secret`)을 비교하고, 조건의 참/거짓 여부에 따라 서로 다른 결과(`too low`, `Correct!`, `Too high`)를 화면에 출력하는 분기 처리를 구현한다.


```python
secret = 5
guess = int(input('Guess a number: '))

if guess <=  4:
    print('too low')
elif guess >= 6:
    print('too high')
else:
    print('Bingo! it is same!')
```

```python
secret =  5
guess = int(input('Guess a number: '))
if guess < secret:
    print('too low')
elif guess == secret:
    print('Correct!')
else:
    print('Too high')
```
