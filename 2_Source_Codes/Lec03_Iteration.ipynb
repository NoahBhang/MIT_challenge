## 1. While Loop

while 문은 조건이 True인 동안 계속 반복 실행된다.

기본 구조

```python
while 조건:
    실행문

````python
from math import factorial

where = input("You are in the Lost Forest. Go left or right? ")
````

##### 사용자가 계속 right를 입력하면 반복이 지속되고, 다른 값을 입력하면 반복문이 종료된다.
#### **핵심개념**
##### while문은 "조건기반 반복"이라고 이해할 수 있다.(True/False)

````python
while where == 'right':
    where = input('You are in the Lost Forest. Go left or right?')
print('You got out of the Lost Forest')
````

#### 재밌는 '게임' 하나 만들어 보자!
##### while과 if가 만나 게임이 종료되거나, 다른 축하의 메시지를 전하거나 하는 조건문을 만들 수 있다.

````python
# 게임 시작 (첫 번째 입력 받기)
where = input("You are in the Lost Forest. Go left or right? ")
# 여기서 where는 변수지 데이터 분석에서 "이 조건에 맞으면 A를, 틀리면 B를 택하라"는 의미의 pandas나 Numpy에서 사용하는 where가 아니다. 또 SQL에서 데이터를 뽑아올 때 특정 조건의 것만 필터링하는 역할을하는 WHERE가 아니다.

# 'right'를 입력하는 동안에는 계속 숲을 헤매며 반복한다.
while where == 'right':
    print("Oops! You walked in a circle and ended up in the same place.")
    where = input('You are in the Lost Forest. Go left or right? ')

# while문을 빠져나온 후, 입력된 값에 따라 다른 결과를 보여줍니다.
if where == 'left':
    print('🚨 Oh no! You encountered a scary monster... GAME OVER 👾')
else: # 여기서 말하는 else는 'left'를 제외한 다른 모든 것들을 말한다.가령 'up'을 입력한 경우.
    print('🎉 Congratulations! You safely got out of the Lost Forest!')
````

#### 다른 문제 한 번 풀어보자.    What is print when you type "right"?

````python
where = input('go left or right? ')
while where == 'right':
    where = input('go right or left?')
print('You got out of the Lost Forest')
````

## **2. Infinite Loop(무한 반복)**

````python
while True:
    print("Noooo")
````

##### 조건이 항상 True이므로 반복이 끝나지 않는다.

##### 실행 중단 방법

######   *  CTRL +C 또는 Command + C
######   * IDE의 Stop 버튼
### *주의점*
##### while문에서는 종료 조건을 반드시 설계해야 한다

##### 무한 루프에 빠지는 경우:  'x'를 무한히 출력하는 이유는 들여쓰기(Indent) 때문이다.
##### *파이썬은 들여쓰기를 기준으로 어디까지가 while문 안에서 반복할 코드인지를 판단한다. 현재 코드는 n = n-1이 while문 바깥에 있어서 n의 값이 줄어들지 않고 있다.*

````python
n = int(input('Enter a non-negative integer: '))
while n > 0:
    print('x')
n = n-1
````

#### 올바르게 멈추도록 수정한 코드
#### 기존 코드: n = n-1이 while과 같은 열에 있다. 이 경우 파이썬은 print('x') 한 줄만 무한히 반복하는 무한 루프에 빠진다.
#### 수정된 코드: print('x')를 실행한 후, n = n-1을 통해 n의 값을 1씩 줄입니다. 결국 n이 0이 되는 순간 while n > 0 조건이 거짓(False)이 되어 반복문이 자연스럽게 멈춘다

````python
n = int(input('Enter a non-negative integer: '))
while n > 0:
    print('x')
    n = n-1  # 💡 이 줄 앞에 들여쓰기(공백 4칸)를 추가했다!
````

## 6. Factorial (팩토리얼)
##### 수학적 정의
###### 5!= 5×4×3×2×1
##### while 버전

````python
factorial = 1
while i  <= x:
    factorial *= i
````

##### for 버전

````python
factorial = 1
for i in range(1, x+1):
    factorial *= i
````

````python
x =4
i =1
factorial = 1
while i <= x:
    factorial *= i
    i += 1
print(f'{x} factroial is {factorial}')
````

````python
x = 4
factorial = 1
for i in range(1, x+1, 1):
    factorial *= i
print(f'{x} factorial is {factorial}')
````

#### **배운점**
##### 반복문을 이용하면

## 7. Summation(누적합)
##### 예시

##### 1. 복합 대입 연산자 사용하기

##### **핵심 패턴**

###### 변수 = 초기값

###### for 반복:
###### <p style="padding-left: 20px;"> 변수 += 값.</p>

###### 이 패턴은 데이터 분석에서도 매우 자주 등장한다
###### 예시
- ###### 총 매출 계산
- ###### 총 주문 건수 계산
- ###### 총 거래 금액 계산


````python
mysum = 0
for i in range(10):
    mysum += i # 혹은 mysum + i
print(mysum)
````

##### 위 예시처럼 *'복합 대입 연산자'* 를 사용할 수 있지만, *'내장 함수 sum()'*  을 사용해 풀 수 도 있다.

##### 2. 내장 함수 sum() 사용하기
##### <p style="padding-left: 20px;">반복문 없이 한 줄로 구현 할 수 있다</p>

````python
mysum = sum(range(10))
print(mysum)
````
