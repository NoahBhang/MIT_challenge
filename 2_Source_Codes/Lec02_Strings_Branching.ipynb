```python
text = input('input Verb: ')

# 1. 입력받은 변수(text)를 문장 사이에 넣어서 출력.
print(f"I can {text} better than you!")

# 2. 변수를 공백(' ')과 함께 5번 반복하여 출력.
print((text + ' ') * 5)
```
